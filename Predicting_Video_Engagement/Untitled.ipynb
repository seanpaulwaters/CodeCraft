```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a multi-directed graph
G = nx.MultiDiGraph()

# Add edges to the graph based on your description
G.add_edge('b', 'c')
G.add_edge('a', 'b')
G.add_edge('a', 'c')
G.add_edge('c', 'a')
G.add_edge('d', 'c')

# Draw the graph
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_size=2000, node_color="skyblue", font_size=15, width=2, edge_color="gray")
plt.show()
```

```python
# Calculate the basic PageRank for all nodes at step k=1
# We'll set the damping factor to 1 (no damping) to get the basic PageRank.
pagerank_values = nx.pagerank(G, alpha=.95)
# Print the PageRank value for node 'c' at step k=1
pagerank_values
```
